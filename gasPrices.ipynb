## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
import pandas as pd
import matplotlib.pyplot as plt


file = 'energy-price.csv'


df = pd.read_csv(file)

selected_cities = ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor', 'London', 'Peterborough', 'St. Catharine\'s', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound']
filtered_data = df.loc[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2023', regex=True)), selected_cities]
mean_prices = filtered_data.mean().sort_values()
plt.figure(figsize=(20, 10))
mean_prices.plot(kind='barh', color='blue')
plt.xlabel('Price in Cents')
plt.ylabel('City')
plt.title('Mean Prices of Regular Unleaded Gasoline in 2023')
plt.grid(axis='x')
plt.show()
```

Plot the corresponding dot plot for the same information presented in Question 1.

## Question 3

Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins.

```python
import pandas as pd
import matplotlib.pyplot as plt

file = 'energy-price.csv'

df = pd.read_csv(file)

toronto_w_data = df.loc[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2023', regex=True))]
plt.figure(figsize=(12,6))
plt.hist(toronto_w_data['Toronto West/Ouest'], bins = 15, color = 'blue', edgecolor = 'black')
plt.title('Histogram of Regular Unleaded Gasoline Prices in Toronto West/Ouest (2023)')
plt.xlabel('Price in Cents')
plt.ylabel('Count')
plt.grid(axis = 'y')
plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
import pandas as pd
import matplotlib.pyplot as plt

file = 'energy-price.csv'

df = pd.read_csv(file)

toronto_w_data = df.loc[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (df['Date'].str.contains('2023', regex=True))]
plt.figure(figsize=(12,6))
plt.ecdf(toronto_w_data['Toronto West/Ouest'])
plt.xlabel('Price')
plt.ylabel('Cumulative Frequency')
plt.title('ECDF (Normalized) for Regular Unleaded Gas Prices in Toronto West/Ouest for 2023')
plt.show()
```
